```python
fig, ax = plt.subplots(3, sharex=True, figsize = (20,8))

lowpass = 100
highpass = 0.01

a,b = signal.butter(6,(highpass,lowpass), btype='bandpass', analog=True)
filtered = signal.lfilter(b,a,data[:1500])
smoothed = signal.cspline1d(filtered, lamb=100)


ax[0].plot(data[:1500])
ax[1].plot(filtered)
ax[2].plot(smoothed)
```

```python
signal_slice = np.ndarray.flatten(smoothed)

peak_idx = peakutils.indexes(signal_slice, thres=0.3, min_dist=200)

fix,ax = plt.subplots(figsize=(20,5))

ax.plot(smoothed)


for peak in peak_idx:

    ax.axvline(x=peak, color='r')
```

```python
peak_idx
```

```python
one_row = smoothed[peak_idx[0]:peak_idx[1]]

fig,ax = plt.subplots(figsize=(20,5))


signal_slice = smoothed[peak_idx[0]+20:peak_idx[0]+140]
rowi = peakutils.indexes(signal_slice, thres=0.3, min_dist=200)

ax.plot(range(peak_idx[0],peak_idx[1]),one_row)

for peak in rowi:
    ax.axvline(x=peak+peak_idx[0]+20, color='r')
```

```python
diff = []

for i,peak in enumerate(peak_idx[:-1]):
    diff.append(peak_idx[i+1] - peak)
diff = np.array(diff)
    
th = np.mean(diff) 
th = th - th*0.25

print(th,len(peak_idx))
to_delete = []
for i,peak in enumerate(peak_idx[:-1]):
    if peak_idx[i+1] - peak < th-th*0.25:
        print(peak, peak_idx[i+1])
        to_delete.append(i+1)

peak_idx = np.delete(peak_idx,to_delete)
len(peak_idx)
```
